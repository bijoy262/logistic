```python
def str_to_list(val):
    if pd.isna(val):
        return []
    if isinstance(val, list):
        return [ast.literal_eval(item) for item in val]
    return ast.literal_eval(val)
list_columns = [
    'description', 'qty', 'qq_title', 'qq_sku', 'supplier_url', 'supplier', 
    'cost_price', 'rate', 'discount', 'discount_type', 'inwFreight', 'final_price'
]
for col in list_columns:
    df[col] = df[col].apply(str_to_list)

# Function to explode the specified columns
def explode_columns(df, list_columns):
    idx_columns = df.columns.difference(list_columns)
    df_exploded = df[list_columns].apply(lambda x: x.apply(pd.Series).stack()).reset_index(level=1, drop=True).reset_index()
    df_exploded = df_exploded.join(df[idx_columns], on='index').drop(columns='index')
    return df_exploded

# Explode the DataFrame
df_exploded = explode_columns(df, list_columns)

# Display the resulting DataFrame
print(df_exploded)
```
